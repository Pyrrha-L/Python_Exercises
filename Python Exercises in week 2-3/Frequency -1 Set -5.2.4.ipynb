```python
class tradingunit:
    '交易单'
    unitct=0
    
    def __init__(self, _no, _movies):
        self.no=_no
        self.movies=_movies
        tradingunit.unitct+=1
    
    def printunit(self):
        print('unit number: ',self.no, '; movie list: ',self.movies)

class movieunit:
    
    def __init__(self, _name):
        self.name=_name
        self.times=0
        self.units=[0]*10
    
    def add(self):
        self.times+=1
    
    def addunit(self,i):
        self.units[self.times]=i
    
    def CalSpt(self,total):
        self.Spt=self.times/total
        
    def printinfo(self):
        print('1.movie name:',self.name, ' 2.appear times:', self.times, ' 3.Supporting Rate:', self.Spt )
        print('trading units: ')
        for i in range(0,self.times):
            print(self.units[i])
        

#交易单初始化
unit1=tradingunit(1,['Six Sense','LOTR1','Harry Potter1','Green Mile','LOTR2'])
unit2=tradingunit(2,['Gladiator','Patriot','Braveheart'])
unit3=tradingunit(3,['LOTR1','LOTR2'])
unit4=tradingunit(4,['Gladiator','Patriot','Six Sense'])
unit5=tradingunit(5,['Gladiator','Patriot','Six Sense'])
unit6=tradingunit(6,['Gladiator','Patriot','Six Sense'])
unit7=tradingunit(7,['Harry Potter1','Harry Potter2'])
unit8=tradingunit(8,['Gladiator','Patriot'])
unit9=tradingunit(9,['Gladiator','Patriot','Six Sense'])
unit10=tradingunit(10,['Six Sense','LOTR1','Gladiator','Green Mile'])

#电影初始化
Sixsense=movieunit('Six Sense')
LOTR1=movieunit('LOTR1')
HP1=movieunit('Harry Potter1')
GM=movieunit('Green Mile')
LOTR2=movieunit('LOTR2')
Gladiator=movieunit('Gladiator')
Patriot=movieunit('Patriot')
Braveheart=movieunit('Braveheart')
HP2=movieunit('Harry Potter2')
LOTR=movieunit('LOTR')

#unit1.printunit()
unitall=[unit1,unit2,unit3,unit4,unit5,unit6,unit7,unit8,unit9,unit10]
movieall=[Sixsense,LOTR1,HP1,GM,LOTR2,Gladiator,Patriot,Braveheart,HP2,LOTR]
moviect=len(movieall)

for i in range (0,tradingunit.unitct):
    for j in range(0,len(unitall[i].movies)):
        for k in range(0,moviect):
            if(unitall[i].movies[j]==movieall[k].name):
                movieall[k].addunit(i+1)
                movieall[k].add()
                break

#过滤输出
for i in range(0,moviect):
    movieall[i].CalSpt(tradingunit.unitct)
    if(movieall[i].Spt>0.5):
        movieall[i].printinfo()
```

```python
#参考资料：《数据科学概论》5.2.1.
#先进行特征选择，再计算其信息增益

import math

def CalculateInfoBef(P):
    ans=0
    for i in range(0,len(P)):
        ans=ans-P[i]*math.log(P[i],2)
    return ans

def CalculateInfoAft(D,InfoD,total): #只进行一次特征选择，分类后信息熵应为0
#D为Dj的数值数组，InfoD为Dj的信息熵数组,total为总集个数|D|
    max=len(D)
    ans=0
    for i in range(0,max):
        ans=ans-(D[i]/total)*InfoD[i]
    return ans

def printresult(A,name):
    print(name,': ')
    for i in range(0,len(A)):
        print(A[i],' ')
    
    print('\n')
    return 0

RawInfo=['young','young','young','young','young',
        'middle','middle','middle','middle','middle',
        'old','old','old','old','old']

youngct=0
middlect=0
oldct=0
count=len(RawInfo)
YoungInfo=[0]*5
MiddleInfo=[0]*5
OldInfo=[0]*5

for i in range(0,count):   #进行特征选择
    if(RawInfo[i]=='young'):
        YoungInfo[youngct]=i+1
        youngct=youngct+1
    elif(RawInfo[i]=='middle'):
        MiddleInfo[middlect]=i+1
        middlect=middlect+1
    else:
        OldInfo[oldct]=i+1
        oldct=oldct+1

printresult(YoungInfo,'young')
printresult(MiddleInfo,'middle')
printresult(OldInfo,'old')

P=[youngct/count,middlect/count,oldct/count]
InfoD=CalculateInfoBef(P)

YoungInfo2=[youngct/youngct]   #计算InfoAD
MiddleInfo2=[middlect/middlect]
OldInfo2=[oldct/oldct]
InfoD2=[CalculateInfoBef(YoungInfo2),CalculateInfoBef(MiddleInfo2),CalculateInfoBef(OldInfo2)]
InfoAD=CalculateInfoAft(P,InfoD2,len(P))
print('InfoAD: ',InfoAD)

Gain=InfoD-InfoAD
print('Gain: ',Gain)
```
